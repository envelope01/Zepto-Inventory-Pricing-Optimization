1) Stock Availability by Category with Visualization

```python
# Group by Category and outOfStock, count the number of products
stock_availability = df.groupby(['Category', 'outOfStock']).size().unstack(fill_value=0)

# Plot stacked bar chart
stock_availability.plot(kind='bar', stacked=True, figsize=(12,7), color=['purple', 'red'])

plt.title('Stock Availability by Category')
plt.xlabel('Category')
plt.ylabel('Number of Products')
plt.xticks(rotation=45, ha='right')
plt.legend(['In Stock (False)', 'Out of Stock (True)'])
plt.tight_layout()
plt.show()
```

### Stock Availability by Category - Insights

From the stacked bar chart, we can see that categories like **Fruits & Vegetables** and **Cooking Essentials** have a large number of products in stock, indicating good availability for customers.  

On the other hand, categories such as **Health & Hygiene** and **Personal Care** show relatively higher out-of-stock counts, which might suggest potential supply issues or higher demand in these categories.


2) What is the average discount percentage across various product categories?

```python
# Calculate average discountPercent per Category
avg_discount = df.groupby('Category')['discountPercent'].mean().sort_values(ascending=False)

# Plot bar chart of average discount percentage by category
plt.figure(figsize=(12,6))
avg_discount.plot(kind='bar', color='purple')

plt.title('Average Discount Percentage by Category')
plt.xlabel('Category')
plt.ylabel('Average Discount (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Average Discount Percentage by Category - Insights

The average discount percentage varies across categories. For example, **Munchies** and **Packaged Food** tend to have higher average discounts, making these categories attractive for budget-conscious customers.

Categories like **Meats, Fish & Eggs** and **Health & Hygiene** show lower average discounts, indicating less frequent or smaller promotions.



3) Which products have the highest discounted selling price, and what categories do they belong to?

```python
# Get top 10 products by discountedSellingPrice
top_discounted_products = df[['name', 'Category', 'discountedSellingPrice']].sort_values(by='discountedSellingPrice', ascending=False).head(10)

print(top_discounted_products)
plt.figure(figsize=(14,6))
plt.barh(top_discounted_products['name'], top_discounted_products['discountedSellingPrice'], color='purple')
plt.xlabel('Discounted Selling Price')
plt.title('Top 10 Most Expensive Discounted Products')
plt.gca().invert_yaxis()  # Highest on top
plt.tight_layout()
plt.show()
```

### Top 10 Most Expensive Discounted Products

The table and bar chart list the top 10 products with the highest discounted selling prices along with their categories.  



4) How does the weight of products vary across different categories?

```python
# Calculate average weight in grams per Category
avg_weight = df.groupby('Category')['weightInGms'].mean().sort_values(ascending=False)

# Plot bar chart of average product weight by category
plt.figure(figsize=(12,6))
avg_weight.plot(kind='bar', color='purple')

plt.title('Average Product Weight by Category (in grams)')
plt.xlabel('Category')
plt.ylabel('Average Weight (grams)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Average Product Weight by Category

This bar chart shows the average weight of products in grams across different categories.  


5) What is the relationship between discount percentage and discounted selling price?

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='discountPercent', y='discountedSellingPrice', alpha=0.6)
plt.title('Discount Percentage vs Discounted Selling Price')
plt.xlabel('Discount Percentage (%)')
plt.ylabel('Discounted Selling Price')
plt.tight_layout()
plt.show()
```

### Relationship Between Discount Percentage and Discounted Selling Price

This scatter plot shows how discount percentages relate to the discounted selling price of products.  
I can observe that products with higher selling prices tend to have varied discount percentages. 


Cooking Essentials and Munchies have the most products, showing strong variety.

Customers can find great deals by focusing on highly discounted products.

Zepto offers some premium products alongside affordable everyday items.

Most products are in stock, but tracking out-of-stock items helps improve inventory.

Categories like Fruits & Vegetables have higher average discounts, boosting customer savings.
